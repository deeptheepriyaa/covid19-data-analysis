```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
df = pd.read_csv("../data/cleaned_covid_data.csv")

# Group by country
top_countries = df.groupby("Country_Region")["Confirmed"].sum().nlargest(10)

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.values, y=top_countries.index, hue=top_countries.index, palette="coolwarm", legend=False)


# Labels
plt.xlabel("Confirmed Cases")
plt.ylabel("Country")
plt.title("Top 10 Countries with Highest COVID-19 Cases")
plt.show()
```
